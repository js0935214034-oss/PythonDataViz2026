### ตัวที่อยู่ข้างใน [ ] เราเรียกว่า index (ตัวชี้)

### List slicing สามารถทำได้โดยใช้ colon :
[a:b] -> [a,b)
9.gif


ในกรอบบนแสดงช่วง $-2 \le x < 3$ หรือเขียนเป็นช่วงได้ว่า [-2, 3)วงเล็บเหลี่ยม [ ข้างหน้า (จุดทึบ 🟢):


แปลว่า "นับรวมตัวนั้นด้วย" (ในทางคณิตศาสตร์คือ $\ge$ หรือ $\le$)วงเล็บโค้ง )


ข้างหลัง (จุดโปร่ง ⚪): แปลว่า "ไม่รวมตัวนั้น" (ในทางคณิตศาสตร์คือ $<$ หรือ $>$)ถ้านำมาเทียบกับ List slicing [a:b]


สมมติมีสไลด์ของข้อมูล [2:5]มันจะเริ่มดึงข้อมูลตั้งแต่ ดัชนี (Index) ที่ 2 ไปเรื่อย ๆแต่จะหยุดลงก่อนถึงตัวสุดท้าย คือดึงถึงแค่ Index ที่ 4 เท่านั้น (ไม่รวม Index ที่ 5)

กรอบล่างแสดงช่วงที่อยู่ตรงกันข้ามกัน คือ $x < -2$ หรือ $x \ge 3$

เขียนได้ว่า (-∞, -2) ∪ [3, ∞)ในทางคณิตศาสตร์ มันคือการแสดงพื้นที่ที่เหลือที่ไม่ได้ถูกเลือกจากกรอบบนถ้าเทียบกับการเขียนโค้ด

ตัวเลขติดลบใน List slicing หมายถึงการ นับจากท้าย List เช่น-1 คือตัวสุดท้าย-2 คือตัวรองสุดท้าย

````python
print(t)
print(t[7:9]) #เอาตัวอักษรตั้งแต่ตัวที่ 7 จนถึงตัวที่ 8 ไม่เอา 9
````

## ให้ สร้าง list ของ เวลา ```['12:30:15','13:41:07',....]``` แล้ววนลูปเรียกฟังก์ชั่นคำนวณเวลาเป็น วินาที

**commit ว่า HW3**

````python
# สร้างฟังก์ชันสำหรับคำนวณเวลาโดยเฉพาะ
def convert_to_seconds(time_str):
    # ใช้ split(':') แยกชั่วโมง นาที วินาที ออกจากกัน
    h, m, s = time_str.split(':')

    # คำนวณเป็นวินาที (แปลงข้อจำเป็น int ก่อน)
    total_seconds = (int(h) * 3600) + (int(m) * 60) + int(s)
    return total_seconds

# สร้าง list ของเวลาที่กำหนด
time_list = [
    '12:30:15',
    '13:41:07',
    '12:53:15',
    '00:59:25',
    '11:11:11',
    '16:06:09'
]
print(f"{'เวลา (HH:MM:SS)':<18} | {'คำนวณเป็นวินาที':<15}")
print("-" * 40)

# วนลูปเพื่อเรียกฟังก์ชันคำนวณและแสดงผล
for time_item in time_list:
    # เรียกใช้งานฟังก์ชันที่สร้างไว้ในขั้นตอนแรก
    seconds = convert_to_seconds(time_item)
    print(f"{time_item:<18} | {seconds:>7,} วินาที")
````

รอบแรกจะหยิบ '12:30:15' มาเก็บไว้ใน time_item แล้วทำคำสั่งด้านล่างให้เสร็จ พอเสร็จแล้วก็จะวนกลับมาหยิบตัวถัดไปคือ '13:41:07' ทำแบบนี้ไปเรื่อย ๆ จนครบทุกตัว

**seconds = convert_to_seconds(time_item)**

1. จะส่งข้อความเวลาที่อยู่ใน
time_item (เช่น '12:30:15') วิ่งเข้าไปในฟังก์ชัน convert_to_seconds

2. ฟังก์ชันจะเอาสตริงนี้ไปตัดแบ่งแตกเป็นวินาที คอยคำนวณจนได้ตัวเลข 45015 ออกมา

3. จากนั้นมันจะโยนตัวเลข 45015 กลับมาเก็บไว้ในตัวแปรที่ชื่อว่า seconds เพื่อเอาไว้ใช้งานต่อ




**print(f"{time_item:<18} | {seconds:>7,} วินาที")**


1. {time_item:<18}: สั่งให้พิมพ์ข้อความเวลาออกมา โดยบังคับว่า "จองพื้นที่ไว้ 18 ตัวอักษรและให้ชิดซ้าย (<)" ถ้าข้อความยาวไม่ถึง มันจะเติมช่องว่างข้างหลังให้เต็ม 18 ช่องอัตโนมัติ

2. |: เครื่องหมายสแลชตรงกลาง พิมพ์ออกมา เพื่อทำเป็นเส้นแบ่งคอลัมน์ตาราง

3. {seconds:>7,}: สั่งให้พิมพ์ตัวเลขวินาทีออกมา โดยมีคำสั่งพิเศษซ่อนอยู่ 2 อย่าง:

      :>7 คือ "จองพื้นที่ไว้ 7 ตัวอักษรและให้ชิดขวา (>)" เพื่อให้หลักหน่วย หลักสิบ หลักร้อย ของตัวเลขตรงกันทุกบรรทัด

      , (เครื่องหมายจุลภาค) คือการสั่งให้ "ใส่คอมมาคั่นหลักพันให้อัตโนมัติ" (เช่น จาก 45015 จะกลายเป็น 45,015)
